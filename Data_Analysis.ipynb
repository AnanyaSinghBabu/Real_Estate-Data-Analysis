```python
# Function to convert price string into number
def convert_price(value):
    if pd.isna(value):  # handle missing
        return None
    value = str(value).replace(",", "").strip()

    if "Crore" in value:
        return float(value.replace("Crore", "").strip()) * 10000000
    elif "Lac" in value:
        return float(value.replace("Lac", "").strip()) * 100000
    else:
        try:
            return float(value)
        except:
            return None

# Apply function
df['price_num'] = df['price'].apply(convert_price)

# Now you can calculate
print("Average Price:", df['price_num'].mean())
print("Max Price:", df['price_num'].max())
```

```python
print(df.groupby("bedRoom")["price_num"].mean())
```

```python
#Price vs Bathrooms
print(df.groupby("bathroom")["price_num"].mean())
```

```python
#Most Expensive Societies
print(df.groupby("society")["price_num"].mean().sort_values(ascending=False).head(10))
```

```python
#Price vs Area (numeric conversion needed)

# Clean area (assuming sqft)
df["area_num"] = df["area"].str.extract(r"(\d+)").astype(float)

# Correlation
print(df[["area_num", "price_num"]].corr())

# Example: average price per sqft
df["price_per_sqft"] = df["price_num"] / df["area_num"]
print(df["price_per_sqft"].mean())
```

```python
# Price distribution
df["price_num"].hist(bins=50)
plt.title("Price Distribution")
plt.show()

# Bedrooms vs Price
df.groupby("bedRoom")["price_num"].mean().plot(kind="bar")
plt.title("Average Price by Bedrooms")
plt.show()

# Area vs Price (scatter plot)
df.plot.scatter(x="area_num", y="price_num", alpha=0.5)
plt.title("Area vs Price")
plt.show()
```

```python
# Price distribution
sns.histplot(df['price_num'], bins=50, kde=True)
plt.title("Price Distribution")
plt.show()

# Bedrooms vs Price
sns.boxplot(x='bedRoom', y='price_num', data=df)
plt.title("Price vs Bedrooms")
plt.show()

# Area vs Price
sns.scatterplot(x='area_num', y='price_num', data=df)
plt.title("Area vs Price")
plt.show()

# Furnishing type counts
sns.countplot(x='furnishDetails', data=df)
plt.title("Furnishing Type Distribution")
plt.xticks(rotation=45)
plt.show()
```
